A distplot is distribution plot. displot is fig or graph that combines the elements of both the histogram and kernal density plot(kde). displots are commonly used in data analysis to visualize the distribution of single variable like histogram.


Here as I am doing the assignment of ploty. I am unable to find distplot in ploty as of now. but its availabe in seaborn and matplotlib.

But as it is a combination of histogram and kde . we can make this as show below

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
fig = go.Figure()
fig_histogram = fig.add_trace(go.Histogram(x=titanic['fare']))
fig_histogram.show()
```

```python
fig = go.Figure()
fig_kde = fig.add_trace(go.Scatter(x=titanic['fare']))
fig_kde.show()
```
