```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y, color = 'red')


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y, color = 'blue')


x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y, color = 'violet')

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y, color = 'yellow')
```

The subplot() function in matplotlib is used to create and arrange multiple subplots within a single figure or canvas. this is useful when we want to display the multiple plots allowing us to compare and visualize different data sets.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2],)
plt.title('Grouped Box Plot')
```

A box plot is a data visualzation technique used to display the distribution of a dataset.
It provides a graphical representation of the variability, central tendency, and potential outliers within a dataset. 

It is use for
Summary of Data
Outlier Detection
Comparison of Distributions
Data Spread

A typical box plot consists of the following components:
Box : It contains the middle 50% of the data.
Median Line : A horizontal line inside the box represents the median of the data
Outlier
