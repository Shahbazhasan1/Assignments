glyphs in Bokeh :

glyphs are the graphical elements used to represent data on the plot. The examples of glyphs are circle, square, traingle, image etc.
glyphs can be customize through various attributes such as color, line style, size etc.

we can do wide range of visualization using bokeh

```python
from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Scatter Plot")

p.scatter(x, y, size=10, color="red", alpha=0.5)

show(p)
```

In this example dataset we use scatter glyph with a attribute color as red with size of 10 and transperancy i.e alpha of o.5

we can customize the appearance of a bokeh plot by using glyph and figure as shown below

1.Title and Title properties
p = figure(title = '',x_axis_label= '', y_axis_label = '', title_text_font_style = '' and so on)

2.Legends customization
p.circle(x, y, size=10, color="blue", legend_label="Data")
p.legend.title = "Legend Title"
p.legend.label_text_font_size = "12pt"

3.plot background and borders by using
p.plot_background_fill_color = "lightgray"
p.border_fill_color = "white"
p.border_line_width = 2

4.fon properties like
p.title.text_font_size = "16pt"
p.xaxis.axis_label_text_font_style = "italic"
p.legend.label_text_font = "helvetica"
