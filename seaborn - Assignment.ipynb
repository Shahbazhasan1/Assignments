```python
penguins = sns.load_dataset('penguins')
```

```python
p = penguins
```

```python
p
```

ex_uses - I want to know that which species has highest and lowest body mass

```python
sns.scatterplot(x = 'species' , y = 'body_mass_g', data = p, )
```

Explanation - Seeing to this graph we can conclude that gento species has highest body mass and chinstrap has the lowest body mass

ex_uses - what is the gender of species which has highest and lowest body mass 

```python
sns.scatterplot(x = 'species' , y = 'body_mass_g', data = p, style = 'sex'  )
```

The species which has highest body mass is male  and which has lowest is female

ex_uses - what is the island of penguin who has max bill depth?

```python
sns.lineplot(x = 'island' , y = 'bill_depth_mm', data = p, hue = 'sex' )
```

ans - torgersen island has the penguin of max bill depth which is male. And we can also conclude that male penguins has high depth of bill comparatively with female
